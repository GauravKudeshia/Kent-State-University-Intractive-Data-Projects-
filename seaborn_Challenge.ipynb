**Q1, Initial Exploration:
Create a scatter plot using Seaborn default settings with Age on the x-axis and Salary on the y-axis.**

```python
plt.figure()
sns.scatterplot(x = 'Age', y = 'Salary', data = salary_data)
plt.title('Default Scatter Plot: Ave vs Salary')
plt.show()
```

**Q2, Visualization:**

- Explore the dataset and seaborn package by creating an advanced scatter plot.
- Include Gender and Education Level as attributes in the legend.
- Customize the color and marker palettes to increase visual appeal.

```python
plt.figure()
sns.scatterplot(x = 'Age',\
               y = 'Salary',\
               hue= 'Education Level',\
               style = 'Gender',\
               data = salary_data)
plt.title('Scatter Plot: Ave vs Salary by Education Level and Gender')
plt.show()
```

```python
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}

# Creating a scatter plot Age vs salary and adding additional information for the education level and gender
plt.figure()
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=salary_data, palette=education_colors)
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

**Q3, Styling and Customization:**

- Set the plot style to remove grid lines for a cleaner look.
- Customize the legend to provide clear information without distracting from the plot.

```python
#Defining custom color palette
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}

#Defining custom color palette
gender_makers={"Female":"o","Male":"s"}
# Creating a scatter plot Age vs salary and adding additional information for the education level and gender
plt.figure()
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=salary_data, palette=education_colors, markers=gender_makers)
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
#Defining custom color palette
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}

#Defining custom color palette
gender_makers={"Female":"o","Male":"s"}

plt.figure(figsize=(12,8))
sns.set_style("ticks")
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=salary_data, palette=education_colors, markers=gender_makers)
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
gender_makers={"Female":"o","Male":"s"}
plt.figure(figsize=(12,8))
sns.set_style("ticks")
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=salary_data, palette=education_colors, markers=gender_makers)

#removing legend frame and dividing it in 2 columns
plt.legend(ncol=2, frameon=False, loc="lower right")
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

**Q4, Additional Analysis :**

- As the final scatter plot is too cluttered, create separate plots for each education level utilizing the filtered data frame option and setting the x and y axes limits.
- Explore other Seaborn plot types (e.g., bar plots, box plots) to extract more insights from the dataset.
 

```python

gender_colors = {"Female": "red", "Male": "blue"}

plt.figure(figsize=(10, 8))
sns.set_style("ticks")  # no gridlines

Bachelors = sns.scatterplot(x="Age", y="Salary", hue="Gender", style="Gender",
                            data=salary_data[salary_data["Education Level"] == "Bachelor's"],
                            palette=gender_colors)

# Setting limits for x and y axes
Bachelors.set_xlim(25, 50)
Bachelors.set_ylim(50000, 250000)

# Customizing the legend
plt.legend(frameon=False, loc="upper left")

plt.title("Age vs Salary Based on Gender Among Bachelor's Degree Holders")
plt.show()
```

```python
gender_colors = {"Female": "red", "Male": "blue"}

plt.figure(figsize=(10, 8))
sns.set_style("ticks")  # no gridlines

Bachelors = sns.scatterplot(x="Age", y="Salary", hue="Gender", style="Gender",
                            data=salary_data[salary_data["Education Level"] == "Master's"],
                            palette=gender_colors)

# Setting limits for x and y axes
Bachelors.set_xlim(25, 50)
Bachelors.set_ylim(50000, 250000)

# Customizing the legend
plt.legend(frameon=False, loc="upper left")

plt.title("Age vs Salary Based on Gender Among Master's Degree Holders")
plt.show()

```

### Creating bar charts in seaborn

```python
#Creating a bar chart
sns.barplot(x='Years of Experience', y = 'Education Level', hue = 'Gender', data=salary_data)
```

```python
sns.boxplot(x='Years of Experience', y='Education Level', hue='Gender', data = salary_data)
plt.show()
```

```python
# Added cap on bar 
sns.barplot(y='Years of Experience',x='Education Level',hue='Gender',capsize=0.2,data=salary_data)
```

```python
#Creating a bar chart with capsize and palette will be dark & blue.
sns.barplot(x='Years of Experience', y = 'Education Level', hue = 'Gender',capsize=0.2, data=salary_data, errorbar=None, palette='dark:blue',hue_order=['Female','Male'])
```
